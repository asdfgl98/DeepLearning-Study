### 이미지 읽고 출력하기

```python
import cv2
```

```python
# 이미지 파일 읽어오기
img = cv2.imread('./img/dog.jpg')
# 이미지 출력
cv2.imshow('dog', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# matplotlib 을 사용하여 이미지 출력
import matplotlib.pyplot as plt
img = cv2.imread('./img/dog.jpg', cv2.IMREAD_COLOR)

plt.imshow(img)
```

- opencv는 색상 공간을 RGB가 아닌 BGR(Blue, Green, Red)로 사용 한다.
- Python 이미지는 RGB로 사용하므로 변환이 필요

```python
img = cv2.imread('./img/dog.jpg', cv2.IMREAD_COLOR)

# 색상변경 옵션 작성
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 그래프 형태의 눈금을 제거해서 사진처럼 출력 ticks
plt.xticks([])
plt.yticks([])

plt.imshow(img)
```

### 컬러이미지의 문제점
- 컬러이미지 3개의 채널로 구성(RGB)
- 시각은 색상정보에 대해 잘못된 정보를 파악하는 경우가 있음
    - 밤에 색상을 볼 때 정확한 색상을 판별하지 못하는 경우, 빛(명도)
    - 채도(흰색)가 달라지는 경우가 있음
- 컬러는 정보를 너무 많이 가지고 있어서 연산량이 많아진다.
    - 속도가 느림
    - 실시간 처리가 어려움
- 컬러이미지를 분석할 때 이진 이미지나 흑백 이미지로 변경해서 분석하는 것이 효율적이다

### Gray(흑백 이미지), Binary(이진 이미지)
- Gray : 0~255 픽셀로 이루어진 이미지 0에 가까울 수록 검정, 255에 가까울수록 흰색
- Binary : 0 또는 1로 구성된 이미지
    - 얼굴만 따로 떼어내는 등 필터링이 필요한 이미지에 사용

#### Gray 이미지 만들기
- 두가지 방법
    1. imread() 를 사용하여 gray 이미지로 읽어오는 방법
    2. 컬러이미지로 불러온 후 gray 이미지로 변경하는 방법

```python
import cv2
import matplotlib.pyplot as plt

# 1. imread() 이미지 불러오기
img = cv2.imread('./img/dog.jpg', cv2.IMREAD_GRAYSCALE) # 이미지를 읽어올 때 흑백으로 읽어오도록 설정
plt.imshow(img, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#2. 컬러이미지로 읽어서 gray 이미지로 변경해주는 방법
img = cv2.imread('./img/dog.jpg', cv2.IMREAD_COLOR)

# 색상변경 옵션 작성
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 그래프 형태의 눈금을 제거해서 사진처럼 출력 ticks
plt.xticks([])
plt.yticks([])

plt.imshow(gray, cmap='gray')
plt.show()
```

### 이진 이미지 만들기
- 이진이미지 : 0과 255로 구성된 이미지
- 주로 관심, 비관심 영역 구분하고 싶을 때 사용

- flag : 문턱값 적용 방법 (다양한 형태의 이진화 방법) 실제로는 위의 두값을 사용
   - cv2.THRESH_BINARY : 픽셀값이 문턱값보다 크면 value, 아니면 0을 할당
  - cv2.THRESH_BINARY_INV : 픽셀값이 문턱값보다 크면 0, 아니면 value를 할당 (inverse)
  - cv2.THRESH_TRUNC : 픽셀값이 문턱값보다 크면 문턱값, 아니면 픽셀값을 그대로 할당
  - cv2.THRESH_TOZERO : 픽셀값이 문턱값보다 크면 픽셀값, 아니면 0을 할당
  - cv2.THRESH_TOZERO_INV : 픽셀값이 문턱값보다 크면 0, 아니면 픽셀값을 할당

```python
# 흑백이미지로 불러오기
gray = cv2.imread('./img/dog.jpg', cv2.IMREAD_GRAYSCALE)

# 이진이미지로 변경
# 문턱 값(임계치) : 기준을 두어 기준보다 작으면 0(뱍) 기준보다 크면 255(흑)으로 변경
# cv2.threshold( 이미지, 문턱값, 문턱값보다 클때 적용하는 값, 적용방법)
# 첫번째 반환 값 : 문턱값(임계치)
# 두번째 반환 값 : 이진이미지
_, thr1 = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)

plt.imshow(thr1, cmap='gray')
plt.show()
```

```python
# inverse(반전)  는 주로 ocr  글씨를 출력할때 사용한다
# 우리가 필요한건 글씨이기때문에 글씨를 검정색으로 출력해주기위함!
plt.subplot(2, 2, 1)  
_, thr1 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thr1, cmap='gray')
plt.title('THRESH_BINARY_INV')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 2)   
_, thr1 = cv2.threshold(gray, 127, 255, cv2.THRESH_TRUNC)
plt.imshow(thr1, cmap='gray')
plt.title('THRESH_TRUNC')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 3)   
_, thr1 = cv2.threshold(gray, 127, 255, cv2.THRESH_TOZERO)
plt.imshow(thr1, cmap='gray')
plt.title('THRESH_TOZERO')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 4)   
_, thr1 = cv2.threshold(gray, 127, 255, cv2.THRESH_TOZERO_INV)
plt.imshow(thr1, cmap='gray')
plt.title('THRESH_TOZERO_INV')
plt.xticks([]), plt.yticks([])
```
